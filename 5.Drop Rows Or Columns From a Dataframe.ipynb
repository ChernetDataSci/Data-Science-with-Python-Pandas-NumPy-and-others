```python
import pandas as pd
```

```python
# reads csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name',inplace = True)
# selecet columns
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

```python
# show dataframe
df
```

# 1 Drop Row 

```python
# drop one row
# axis parameter
df.drop('L. Messi', axis=0)
# index parameter
df.drop(index=['L. Messi'])
```

```python
# drop two or more rows and update a data (inplace = True)
df.drop(['L. Messi', 'Cristiano Ronaldo'], axis=0, inplace = True)
```

```python
df
```

# 2 Drop Columns 

```python
# drop one column
# axis parameter
#df.drop('long_name', axis=1)
# columns parameter
df.drop(columns=['long_name'])
```
