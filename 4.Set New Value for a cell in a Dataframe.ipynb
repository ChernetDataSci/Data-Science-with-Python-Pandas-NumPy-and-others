```python
import pandas as pd
```

```python
# reads csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name',inplace = True)
# selecet columns
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

```python
# show Dataframe
df
```

# 1 Set value to one cell

```python
# update L.Messi height
df.loc['L. Messi','height_cm'] = 175
```

```python
# Show dataframe
df
```

# 2 Set value to entire column

```python
# set height of all player to 190
df.loc[:, 'height_cm'] = 190
```

```python
# show dataframe 
df
```

# 3 Set value to entire row

```python
# get all column that correspond to player ranked last in fifa
import numpy as np
df.iloc[-1, :] = np.nan
```

```python
# set null values to all columns that corresponds to player ranked last 
df.iloc[-1, :] = 0
```

```python
df
```

# 4 Set value to Multiple cells

```python
# Set value for all items matching the list of labels
df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']] = 175
```

```python
# show dataframe
df
```

# 5 Set value for rows matching a condition 

```python
# set value for rows matching a condition 
columns = ['age', 'dob', 'height_cm', 'weight_kg']
df.loc[df['height_cm'] > 180,columns] = 0
```

```python
# show dataframe
df
```
