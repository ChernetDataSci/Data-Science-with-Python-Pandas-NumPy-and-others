# 1 The re module 

```python
import re
```

# 1.1 re.search()

```python
text = """^Hello Wold.
I Love Python!!!!!
My First Lucky number is 987-654-321
My second Lucky number is 876-543-210
My third Lucky number is 765-432-100
My Fourth Lucky number is 123-456-123-123-456
"""
```

```python
# return the first match
re.search(r'\d', text)
```

```python
# flags
re.search(r'hello', text, flags=re.M)
```

# 2 re findall()

```python
# return all the matches
re.findall(r'\d', text)
```

# Exercise 

```python
# Match only punctuation
text = """Hello Wold.
I Love Python!!!!!
My First Lucky number is 987-654-321
My second Lucky number is 876-543-210
My third Lucky number is 765-432-100
My Fourth Lucky number is 123-456-123-123-456
"""
re.findall(r'[^w\s]', text)
```

```python
# Match right date format (MM/DD/YYYY)
text = '''
13-04-2021
2021-13-04
2021-04-13
'''
re.findall(r'\d{2}-\d{2}-\d{4}', text)  
```

```python
# Match right username format (4-14 characters and only letters or digit)
text = '''
username10
abc
10
'''
re.findall(r'[a-zA-Z0-9]{4,14}', text)
```
