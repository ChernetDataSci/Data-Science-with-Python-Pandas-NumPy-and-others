```python
import pandas as pd
```

```python
# reads csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
#show dataframe
df
```

# 1 copy()

# 1.1 deep = True

```python
# deep=True by default (Modifications to the data or indice of the copy )
df_copy = df.copy()
```

```python
# update value in original dataframe
df.loc['L. Messi','height_cm'] = 180
```

```python
# copy vs orginal dataframe
df
df_copy
```

# 1.2 deep = False

```python
# deep=false (any changes to the data of the orginal will be reflected)
df_shallow_copy = df.copy(deep=False)
```

```python
# update value in orginal dataframe
df.loc['Cristiano Ronaldo', 'height_cm'] = 200
```

```python
# shallow copy vs orginal dataframe
df
df_shallow_copy
```

# 1.3 Simple assignment

```python
# make a copy
df_new_copy = df
```

```python
# update value in orginal dataframe
df.loc['Neymar Jr','height_cm'] = 190
```

```python
# copy vs orginal dataframe
df
df_new_copy 
```
